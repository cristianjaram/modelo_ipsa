```python
import pandas as pd
import numpy as np
```

```python
# Lectura de los datos del ipsa

#ruta_ipsa = 'datos_csv_iniciales/Datos_historicos_SP_CLX_IPSA.csv'
#ipsa = pd.read_csv(ruta_ipsa)
ipsa = pd.read_csv("Datos_historicosIPSA.csv")

# Lectura datos de cencosud
#ruta_cencosud = 'datos_csv_iniciales/Datos_historicos_CENCOSUD.csv'
#cencosud = pd.read_csv(ruta_cencosud)
cencosud = pd.read_csv("Datos_historicos_CENCOSUD.csv")
# Lectura datos de copec
#ruta_copec = 'datos_csv_iniciales/Datos_historicos_COPEC.csv'
#copec = pd.read_csv(ruta_copec)
copec = pd.read_csv("Datos_historicos_COPEC.csv")
# Lectura datos de enelam
#ruta_enelam = 'datos_csv_iniciales/Datos_historicos_ENELAM.csv'
#enelam = pd.read_csv(ruta_enelam)
enelam = pd.read_csv("Datos_historicos_ENELAM.csv")
# Lectura datos de futuros cobre (CU) 
#ruta_cu = 'datos_csv_iniciales/Datos_historicos_Futuros_cobre.csv' 
#cu = pd.read_csv(ruta_cu)
cu = pd.read_csv("Datos_historicos_Futuros_cobre.csv")
# Lectura datos de usd vs clp (USD)
#ruta_usd = 'datos_csv_iniciales/Datos_historicos_USD_CLP.csv' 
#usd = pd.read_csv(ruta_usd)
usd = pd.read_csv("Datos_historicos_USD_CLP.csv")
```

```python
# Creacion de las columanas necesarias en la matriz de caracteristicas
ipsa['dia'] = np.nan
ipsa['mes'] = np.nan
ipsa['anho'] = np.nan
ipsa['var_dos_dias'] = np.nan
ipsa['var_tres_dias'] = np.nan
ipsa['var_semana'] = np.nan
ipsa['var_empresa_uno'] = 0
ipsa['var_empresa_dos'] = 0
ipsa['var_empresa_tres'] = 0
ipsa['dolar'] = 0
ipsa['cu'] = 0
ipsa['subio'] = np.nan

# Cambio de nombre de columnas ya presentes + creacion de la version 2 de la matriz
ipsa_v2 = ipsa.rename(columns={'Fecha':'fecha', 'Último':'cierre', 'Apertura':'apertura', 'Máximo':'max', 'Mínimo':'min', '% var.':'var_un_dia'})
```

```python
# Visualizacion de la matriz para ver que se hayan efectuado todos los cambios deseados 
ipsa_v2.head()
```

```python
# Aqui se le define la variable subio a cada columna, esto se hace iterando entre las filas de la matriz y comparando
# los valores de cierre y apertura, si la diferencia es positiva entonces se le asigna el valor 'True' a la columna 
# subio, en el caso contrario se asigna el valor 'False'
for indice, fila in ipsa_v2.iterrows():
    i = 1
    cierre = float((fila['cierre'].replace('.','')).replace(',','.'))
    apertura  = float((fila['apertura'].replace('.','')).replace(',','.'))
    diferencia = cierre - apertura
    if diferencia > 0:
        ipsa_v2.loc[indice, ['subio']] = True
    else:
        ipsa_v2.loc[indice, ['subio']] = False      
```

```python
# Se ven las columnas finales de la matriz para ver en donde termina y verificar si se definio bien
# la columna subio
ipsa_v2.tail()
```

```python
# Se rellenan los valores de las columnas dia, mes y anho iterando sobre las filas de la matriz
for indice, fila in ipsa_v2.iterrows():
    ipsa_v2.loc[indice, ['dia']] = ((ipsa_v2['fecha'])[indice])[0] + ((ipsa_v2['fecha'])[indice])[1]
    ipsa_v2.loc[indice, ['mes']] = ((ipsa_v2['fecha'])[indice])[3] + ((ipsa_v2['fecha'])[indice])[4]
    ipsa_v2.loc[indice, ['anho']] = ((ipsa_v2['fecha'])[indice])[8] + ((ipsa_v2['fecha'])[indice])[9] 
```

```python
# Visualizar el head de la matriz para ver si es correcto el codigo o no
ipsa_v2.head()
```

```python
# Iterar sobre las filas de la matriz para correr los valores de cierre en una fila, esto es porque para predecir si hoy sube 
# el precio del IPSA necesitamos ocupar los valores del dia anterior, esto se va a realizar con todos los valores 
# se comienza con el de cierre para verfificar que el codigo utilizado sea correcto, para despues usarlo con los otros valores
for indice, fila in ipsa_v2.iterrows():
    if indice >= 1242:
        break
    else:
        ipsa_v2.loc[indice, ['cierre']] = (ipsa_v2['cierre'])[indice + 1]
```

```python
# Iterer sobre las filas de la matriz para correr los valores de apertura, max, min y var_un_dia
for indice, fila in ipsa_v2.iterrows():
    if indice >= 1242:
        break
    else:
        ipsa_v2.loc[indice, ['apertura']] = (ipsa_v2['apertura'])[indice + 1]
        ipsa_v2.loc[indice, ['max']] = (ipsa_v2['max'])[indice + 1]
        ipsa_v2.loc[indice, ['min']] = (ipsa_v2['min'])[indice + 1]
        ipsa_v2.loc[indice, ['var_un_dia']] = (ipsa_v2['var_un_dia'])[indice + 1]
```

```python
# Se revisa el head de la matriz para ver que los cambios hayan sido efectivos 
ipsa_v2.head()
```

```python
# Se cambia el formato de los datos a nomenclatura de python para los numeros, iterando sobre las filas de la matriz
for indice, fila in ipsa_v2.iterrows():
    ipsa_v2.loc[indice, ['cierre']] = (ipsa_v2['cierre'][indice].replace('.','')).replace(',','.')
    ipsa_v2.loc[indice, ['apertura']] = (ipsa_v2['apertura'][indice].replace('.','')).replace(',','.')
    ipsa_v2.loc[indice, ['max']] = (ipsa_v2['max'][indice].replace('.','')).replace(',','.')
    ipsa_v2.loc[indice, ['min']] = (ipsa_v2['min'][indice].replace('.','')).replace(',','.')
    ipsa_v2.loc[indice, ['var_un_dia']] = (ipsa_v2['var_un_dia'][indice].replace('%','')).replace(',','.')
```

```python
# Se revisa el head de la matriz para ver que se haya cambiado el formato de los datos efectivamente 
ipsa_v2.head()
```

```python
# Se utiliza la funcion de pandas .to_numeric sobre todas las columnas que contengan datos numericos 
ipsa_v2['dia'] = pd.to_numeric(ipsa_v2['dia'], errors='coerce')
ipsa_v2['mes'] = pd.to_numeric(ipsa_v2['mes'], errors='coerce')
ipsa_v2['anho'] = pd.to_numeric(ipsa_v2['anho'], errors='coerce')
ipsa_v2['cierre'] = pd.to_numeric(ipsa_v2['cierre'], errors='coerce')
ipsa_v2['apertura'] = pd.to_numeric(ipsa_v2['apertura'], errors='coerce')
ipsa_v2['max'] = pd.to_numeric(ipsa_v2['max'], errors='coerce')
ipsa_v2['min'] = pd.to_numeric(ipsa_v2['min'], errors='coerce')
ipsa_v2['var_un_dia'] = pd.to_numeric(ipsa_v2['var_un_dia'], errors='coerce')
```

```python
# Revisamos que en efecto se hayan cambiado todos los valores a datos del tipo float
ipsa_v2.dtypes
```

```python
# Se calcula las variaciones de 2, 3 y 7 dias iterando sobre todas las filas, teniendo cuidado con las ultimas columnas
# en las que no se van a poder calcular alguno de estos valores 
for indice, fila in ipsa_v2.iterrows():
    if indice + 7 > 1242:
        break
    else:
        ipsa_v2.loc[indice ,['var_dos_dias']] = np.round(100*(ipsa_v2['cierre'][indice] - ipsa_v2['apertura'][indice + 2])/(ipsa_v2['apertura'][indice + 2]),2)
        ipsa_v2.loc[indice ,['var_tres_dias']] = np.round(100*(ipsa_v2['cierre'][indice] - ipsa_v2['apertura'][indice + 3])/(ipsa_v2['apertura'][indice + 3]),2)
        ipsa_v2.loc[indice ,['var_semana']] = np.round(100*(ipsa_v2['cierre'][indice] - ipsa_v2['apertura'][indice + 7])/(ipsa_v2['apertura'][indice + 7]),2)
```

```python
# Se revisa el head de la matriz para ver que se hayan definido bien los valores 
ipsa_v2.head()
```

```python
# Se elimnan las ultimas columnas de la matriz en donde hay valores que no estan bien definidos y se crea la tercera version
# de la matriz.
ipsa_v3 = ipsa_v2[:-13]
```

```python
# Se formatean los datos de variacion de todos los otros archivo csv para que sean numeros, ademas, analogamente con el caso de la matriz 
# anterior se van a correr todos los datos de las otras matrices en un dia.

for indice, fila in copec.iterrows():
    if indice >= 1247:
        break
    else:
        copec.loc[indice, ['% var.']] = (str((copec['% var.'])[indice + 1]).replace('%', '')).replace(',','.')
        
copec['% var.'] = pd.to_numeric(copec['% var.'], errors='coerce')

for indice, fila in cencosud.iterrows():
    if indice >= 1245:
        break
    else:
        cencosud.loc[indice, ['% var.']] = (str((cencosud['% var.'])[indice + 1]).replace('%', '')).replace(',','.')
        
cencosud['% var.'] = pd.to_numeric(cencosud['% var.'], errors='coerce')

for indice, fila in enelam.iterrows():
    if indice >= 1248:
        break
    else:
        enelam.loc[indice, ['% var.']] = (str((enelam['% var.'])[indice + 1]).replace('%', '')).replace(',','.')
        
enelam['% var.'] = pd.to_numeric(enelam['% var.'], errors='coerce')

for indice, fila in cu.iterrows():
    if indice >= 1311:
        break
    else:
        cu.loc[indice, ['% var.']] = (str((cu['% var.'])[indice + 1]).replace('%', '')).replace(',','.')
        
cu['% var.'] = pd.to_numeric(cu['% var.'], errors='coerce')

for indice, fila in usd.iterrows():
    if indice >= 1304:
        break
    else:
        usd.loc[indice, ['% var.']] = (str((usd['% var.'])[indice + 1]).replace('%', '')).replace(',','.')
        
usd['% var.'] = pd.to_numeric(usd['% var.'], errors='coerce')

```

```python
# Ahora para los datos de copec se compara la fecha de los datos de copec con la fecha de los datos del ipsa y si la fecha coincide 
# se hace corresponder el dato de copec con la variacion de la empresa uno, hace en un inicio solo con copec para verficar que el 
# codigo empleado sea correcto.
for indice, fila in ipsa_v3.iterrows():
    fecha = ipsa_v3['fecha'][indice]
    for indice_copec, fila_copec in copec.iterrows():
        if fecha == copec['Fecha'][indice_copec]:
            var_empresa_uno = copec['% var.'][indice_copec]
            break
        else:
            pass
    ipsa_v3.loc[indice, ['var_empresa_uno']] = var_empresa_uno   
```

```python
# Se ve el head de la matriz para verificar que esten bien correspondidos los datos
ipsa_v3.head()
```

```python
# Se hace lo mismo que en el caso anterior pero ahora con todos los otros datos faltantes que se obtienen de los otros archivos.
for indice, fila in ipsa_v3.iterrows():
    fecha = ipsa_v3['fecha'][indice]
    for indice_cencosud, fila_cencosud in cencosud.iterrows():
        if fecha == cencosud['Fecha'][indice_cencosud]:
            var_empresa_dos = cencosud['% var.'][indice_cencosud]
            break
        else:
            pass
    ipsa_v3.loc[indice, ['var_empresa_dos']] = var_empresa_dos  
    
    for indice_enelam, fila_enelam in enelam.iterrows():
        if fecha == enelam['Fecha'][indice_enelam]:
            var_empresa_tres = enelam['% var.'][indice_enelam]
            break
        else:
            pass
    ipsa_v3.loc[indice, ['var_empresa_tres']] = var_empresa_tres  
    
    for indice_cu, fila_cu in cu.iterrows():
        if fecha == cu['Fecha'][indice_cu]:
            var_cu = cu['% var.'][indice_cu]
            break
        else:
            pass
    ipsa_v3.loc[indice, ['cu']] = var_cu
    
    for indice_usd, fila_usd in usd.iterrows():
        if fecha == usd['Fecha'][indice_usd]:
            var_usd = usd['% var.'][indice_usd]
            break
        else:
            pass
    ipsa_v3.loc[indice, ['dolar']] = var_usd
```

```python
# Se revisa que se hayan colocado bien los datos
ipsa_v3.head()
```

```python
# Eliminar las filas a las que no les haya corresponido ningun dato
for indice, fila in ipsa_v3.iterrows():
    var_empresa_uno = ipsa_v3['var_empresa_uno'][indice]
    var_empresa_dos = ipsa_v3['var_empresa_dos'][indice]
    var_empresa_tres = ipsa_v3['var_empresa_tres'][indice]
    var_cu = ipsa_v3['cu'][indice]
    var_usd = ipsa_v3['dolar'][indice]
    if var_empresa_uno == 0 or var_empresa_dos == 0 or var_empresa_tres == 0 or var_cu == 0 or var_usd == 0:
        print(str(indice)) # Ver que filas se eliminaron
        ipsa_v3.drop(indice,
                     axis=0,
                     inplace = True)
    else:
        pass
```

```python
# Exportar la matriz ya terminada
ipsa_v3.to_csv('ipsa_matriz_final.csv')
```
